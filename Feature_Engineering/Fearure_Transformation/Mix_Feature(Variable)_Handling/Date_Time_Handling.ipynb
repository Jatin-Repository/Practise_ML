## Working with Date

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

### Extract Year from Date

```python
date['date_year'] = date['date'].dt.year   # dt is datetime(dt)
```

```python
date.sample(10)
```

```python
date['date_month_in_number'] = date['date'].dt.month
```

```python
date.sample(5)
```

```python
date['date_month_name'] = date['date'].dt.month_name()
```

```python
date.sample(5)
```

```python
date['date_day'] = date['date'].dt.day
```

```python
date.sample(5)
```

```python
date['date_day_name'] = date['date'].dt.day_name()
```

```python
date.sample(5)
```

```python
date['date_day_of_the_week'] = date['date'].dt.day_of_week
```

```python
date.sample(5)
```

```python
date.drop(columns=['product_id','city_id','orders']).head()
```

```python
date['date_is_weekend'] = np.where(date['date_day_name'].isin(['Sunday','Saturday']),1,0)
```

```python
date.sample(5)
```

## Extract week in year

```python
date['date_week'] = date['date'].dt.isocalendar().week

date.drop(columns=['product_id','city_id','orders']).head()
```

## Extract Quater

```python
date['day_in_quater'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

## Semester

```python
date['day_in_quater'] = np.where(date['day_in_quater'].isin([1,2]),1,2)

date.drop(columns=['product_id','city_id','orders']).head()
```

## Extract Time Elapsed between Dates

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today -date['date']
```

```python
(today -date['date']).dt.days
```

```python
np.round((today -date['date']) / np.timedelta64(1, 'M'),0)  # month passed
```
