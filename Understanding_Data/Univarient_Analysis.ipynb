## b. Distplot
      1. It is use to identify skewness in data or not.
      2. Positive Skew : People vs Salary (y vs x axis), very less people have more salary.
      3. Negative Skew : Easy Exam vs Marks(y vs x axis), very less student has less mark, majority score good mark.


```python
sns.displot(df['Age'])
```

```python

sns.rugplot(data=df['Age'])
```

```python
sns.kdeplot(data=df,x=df['Age']) # show pdf, also check skewness
```

```python
sns.displot(df['Age'],kde=True) #Kernal Denstity Estimation (kde)
```

## c. Boxplot



    1. The larger the Inter Quartile Range(IQR) the larger variation across the data.
    2. a. Minimum = Q1-1.5*IQR , Maximum = Q3+1.5*IQR
       b. Q1(Lower Quartile) = 25th Percenticle, Q2 = 50th Percentile(Median),Q3(Upper Quartile)= 75th Percentile.
    3. Generally use when dealing with noisy data.

```python
sns.boxplot(df['Fare'],orient='v')
```

```python
sns.boxplot(df['Fare'],orient='h') 
# black dots indicate outlier, relaying on it will be very difficult for model to predict.
```

```python
sns.boxplot(df['Age'],orient='v')
```
