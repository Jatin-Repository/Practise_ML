```python
headers = {"User-Agent":"Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36"}
requests.get('https://www.glassdoor.co.in/Reviews/index.htm',headers=headers)
```

```python
webpage = requests.get('https://www.glassdoor.co.in/Reviews/index.htm',headers=headers).content
```

```python
# Reference: https://github.com/chris-greening/instascrape/pull/150/commits/75b5ff069ebf486df45e4d768b9ec6b24b2b44b0
#soup = BeautifulSoup(webpage, features="xml")
soup = BeautifulSoup(webpage, features="html.parser")
```

```python
soup
```

```python
soup.find_all('h1')[0]
```

```python
for i in soup.find_all('h2'):
    print(i.text)
```

```python
for i in soup.find_all('b'):
    print(i.text)
```

```python
soup.find_all('span', {'class' : 'pr-xsm ratingsWidget__RatingsWidgetStyles__rating'})
```

```python
for i in soup.find_all('span', {'class' : 'pr-xsm ratingsWidget__RatingsWidgetStyles__rating'}):
    print(i.text)
```
