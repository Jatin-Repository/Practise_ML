```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,Y = make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y)
```

```python
from sklearn.linear_model import LinearRegression # apply OLS
```

```python
reg = LinearRegression()
reg.fit(X,Y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Let apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the satring value of intercept b = 0
# b = 0 # we can cahnge the value of b and see other outcomes also
b = 100
b_old = b
Y_pred = ((78.35*X)+b)
Y_pred
```

```python
Y_pred = Y_pred.reshape(4)
Y_pred
```

```python
plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='Ordinary Least Square (OLS)')
plt.plot(X,Y_pred,color='#00a65a',label='Intercept (b = {})'.format(b_old))
plt.legend()
plt.show()
```

```python
Y
```

```python
m = 78.35
# n* slope is step size
loss_slope = -2 * np.sum(Y - (m * X.ravel()) - b)
loss_slope
```

```python
# assume learning rate is 0.1
lr = 0.1

step_size = loss_slope*lr
step_size
```

```python
b = b -step_size
b # b_new
```

```python
b_1 = b
```

```python
Y_pred_1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='Ordinary Least Square (OLS)')
plt.plot(X,Y_pred_1,color='#00a65a',label='Intercept (b={})'.format(b_1))
plt.plot(X,Y_pred,color='#A3ED47',label='Intercept (b= {})'.format(b_old))
plt.legend() # for label printing
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum( Y - m*X.ravel() - b)
loss_slope
```

```python
step_size  = loss_slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
b_2 = b
```

```python
Y_pred_2 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='Ordinary Least Square (OLS)')
plt.plot(X,Y_pred_2,color='#00a65a',label='Intercept (b={})'.format(b_2))
plt.plot(X,Y_pred_1,color='#A3E4D7',label='Intercept (b={})'.format(b_1))
plt.plot(X,Y_pred,color='#A3ED47',label='Intercept (b={})'.format(b_old))
plt.legend() # for label printing
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum( Y - m*X.ravel() - b)
loss_slope
```

```python
step_size  = loss_slope * lr
step_size
```

```python
b = b - step_size
b
```

```python
b_3 = b
```

```python
Y_pred_3 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='Ordinary Least Square (OLS)')
plt.plot(X,Y_pred_3,color='#00a65a',label='Intercept (b={})'.format(b_3))
plt.plot(X,Y_pred_2,color='#A3E4D7',label='Intercept (b={})'.format(b_2))
plt.plot(X,Y_pred_1,color='#A3E4D7',label='Intercept (b={})'.format(b_1))
plt.plot(X,Y_pred,color='#A3ED47',label='Intercept (b={})'.format(b_old))
plt.legend() # for label printing
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.1

epochs = 10

for i in range(epochs):
    loss_slope = -2 * np.sum(Y - m*X.ravel() - b)
    b = b - (lr * loss_slope)

    Y_pred = m * X + b

    plt.plot(X,Y_pred)
plt.scatter(X,Y)
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.01

epochs = 10

for i in range(epochs):
    loss_slope = -2 * np.sum(Y - m*X.ravel() - b)
    b = b - (lr * loss_slope)

    Y_pred = m * X + b

    plt.plot(X,Y_pred)
plt.scatter(X,Y)
plt.show()
```
