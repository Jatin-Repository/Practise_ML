```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
centriods = [(-5,-5),(5,5)]
cluster_std = [1,1]

X,Y = make_blobs(n_samples=100,cluster_std=cluster_std,centers=centriods,n_features=2,random_state=2)
```

```python
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
import random
import numpy as np
class KMeans:
    def __init__(self,n_clusters=2,max_iter=100):
        self.n_clusters=n_clusters
        self.max_iter = max_iter
        self.centroids = None
    
    def fit_predict(self,X):
        #print(X)
        row_random_index = random.sample(range(0,X.shape[0]),self.n_clusters) # any n_cluster index
        self.centroids = X[row_random_index]
        #print(self.centroids)
        for i in range(self.max_iter):

            # assign cluster
            cluster_group = self.assign_clusters(X)
            old_centroids = self.centroids

            #print(old_centroids)
            #move centriod
            self.centroids = self.move_centriods(X,cluster_group)

            #print(self.centroids)
            # check finish
            if np.array_equal(old_centroids,self.centroids):
                break
            # if (old_centroids == self.centroids).all(): # array comparision
            #     break
            
        return cluster_group
    
    def assign_clusters(self,X):
        cluster_group = []

        #distances = []

        for row in X:
            distances = []
            for centriod in self.centroids:
                distances.append(np.sqrt(np.dot(row-centriod,row-centriod)))
            #print(distances)
            #print('------------------------')
            min_distances = min(distances)
            index_pos = distances.index(min_distances)
            cluster_group.append(index_pos)
        return np.array(cluster_group)
    

    def move_centriods(self,X,cluster_group):
        new_centriods = []

        cluster_type = np.unique(cluster_group)
        for type in cluster_type:
            new_centriods.append(X[cluster_group==type].mean(axis=0))# column wise
        
        return new_centriods
    

```

```python
km = KMeans(n_clusters=2,max_iter=100)
Y_mean = km.fit_predict(X)
plt.scatter(X[Y_mean == 0,0],X[Y_mean == 0,1], color = 'red')
plt.scatter(X[Y_mean == 1,0],X[Y_mean == 1,1], color = 'blue')
plt.show()
```
