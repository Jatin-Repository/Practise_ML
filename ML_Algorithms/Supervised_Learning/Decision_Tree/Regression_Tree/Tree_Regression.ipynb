```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
```

```python
boston = pd.read_csv('HousingData.csv')
boston.head()
```

```python
boston.describe()
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
columns = boston.columns
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),columns),
],remainder='passthrough')
```

```python
t = transformer.fit_transform(boston)
new_df = pd.DataFrame(t,columns=columns)
new_df = np.round(new_df,2)
new_df
```

```python
X = new_df.iloc[:,0:-1]
Y = new_df.iloc[:,-1]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
reg_tree = DecisionTreeRegressor(criterion='squared_error',max_depth=5)
```

```python
reg_tree.fit(X_train,Y_train)
```

```python
reg_tree.__dict__
```

```python
Y_pred = reg_tree.predict(X_test)
```

```python
r2_score(Y_test,Y_pred)
```

## Hyperparamter Tunning

```python
param_grid = {
    'max_depth':[2,4,8,10,None],
    'criterion':['squared_error','absolute_error'],
    'max_features':[0.25,0.5,1.0],
    'min_samples_split':[0.25,0.5,1.0]
}
```

```python
reg_tree_GSCV = GridSearchCV(DecisionTreeRegressor(),param_grid=param_grid)
```

```python
reg_tree_GSCV.fit(X_train,Y_train)
```

```python
reg_tree_GSCV.__dict__
```

```python
Y_pred = reg_tree_GSCV.predict(X_test)
r2_score(Y_test,Y_pred)
```

```python
reg_tree_GSCV.best_estimator_
```

```python
reg_tree_GSCV.best_params_
```

## Feature Importance

```python
for importance, name in sorted(zip(reg_tree.feature_importances_,X_train.columns),reverse=True):
    print(name,importance)
```
