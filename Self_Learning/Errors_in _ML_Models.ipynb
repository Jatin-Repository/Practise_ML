```python
from sklearn.datasets import make_regression
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
```

```python
X,Y = make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=42)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train,X_val,Y_train,Y_val = train_test_split(X_train,Y_train,test_size=0.2,random_state=3)
```

```python
X_train.shape
```

```python
X_val.shape
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,X_train)

Y_val_pred = lr.predict(X_val)

print(Y_val_pred.flatten())
```

```python
Y_val_pred =Y_val_pred.ravel()
```

```python
print('Validation Error:',np.sqrt(mean_squared_error(Y_val,Y_val_pred.ravel())))
```

```python
Y_test_pred = lr.predict(X_test)
print('Testing Error:',np.sqrt(mean_squared_error(Y_test,Y_test_pred.ravel())))
```

```python
import matplotlib.pyplot as plt
plt.plot(X_train,Y_train,'b.',label="Training Data Points")
plt.plot(X_val,Y_val,'r.',label="Validation Data Points")
plt.plot(X_test,Y_test,'g.',label= "Testing Data Points")
plt.legend()
plt.show()
```

```python
plt.plot(X,Y,'b.')
#plt.plot(X_train,Y_train,'b.',label="Training Data Points")
plt.plot(X_val,Y_val,'r.',label="Validation Data Points")
plt.plot(X_test,Y_test,'g.',label= "Testing Data Points")
plt.plot(X,lr.predict(X),color='pink',label='Best Fit Line')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import Ridge # indicate model was overfiited as we have manually introduce error in it.

rr1 = Ridge(alpha=0.5)

rr1.fit(X_train,Y_train)

Y_val_pred = rr1.predict(X_val)

Y_test_pred = rr1.predict(X_test)

print('Validation Error using Ridge:',np.sqrt(mean_squared_error(Y_val,Y_val_pred.ravel())))
print('Testing Error using Ridge:',np.sqrt(mean_squared_error(Y_test,Y_test_pred.ravel())))
```
